```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from tqdm import tqdm
```

<h1> Load the Dataset</h1>

```python
transform = transforms.Compose([
    transforms.ToTensor()
])


#load train dataset
train_dataset = datasets.CIFAR10(root = "./data", train = True, transform = transform, download = True)

#load test dataset
test_dataset = datasets.CIFAR10(root = "./data", train = False, transform = transform, download = True)
```

```python
batch_size = 64

#loaders
train_loader = DataLoader(train_dataset, batch_size = batch_size, shuffle = True)
test_loader = DataLoader(test_dataset, batch_size = batch_size, shuffle = True)
```

```python
num_classes = len(train_dataset.classes)
num_classes
```

<h1>Model Development </h1>

```python
class CNN2D(nn.Module):
  def __init__(self, num_classes):
    super().__init__()
    self.conv_layers = nn.Sequential(
        nn.Conv2d(in_channels = 3, out_channels = 16, kernel_size = 3, padding = 1, stride = 1),
        nn.ReLU(),
        nn.MaxPool2d(2),
        nn.Conv2d(16, 32, kernel_size = 3, padding = 1, stride = 1),
        nn.ReLU(),
        nn.MaxPool2d(2),
        nn.Conv2d(32, 64, kernel_size = 1, padding = 0),
        nn.ReLU(),
        nn.Conv2d(64, 128, kernel_size=1, padding = 0),
        nn.Flatten(),
       )

    #fc
    self.fc =  nn.Linear(8192, num_classes)

  #forward pass
  def forward(self, x):
    x = self.conv_layers(x)
    x = self.fc(x)
    return x
```

```python
model = CNN2D(num_classes=num_classes)
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(params = model.parameters(), lr = 0.001)
```

```python
n_epochs = 20
loss_arr = []


for epoch in range(n_epochs):
  #loader
  loss_b = 0.0
  for x_batch, y_batch in tqdm(train_loader):
    optimizer.zero_grad()
    #forward pass
    y_pred = model(x_batch)

    #calc loss
    loss = criterion(y_pred, y_batch)
    loss_b += loss.item()

    #backprop
    loss.backward()
    optimizer.step()


  #epoch end
  print(f'Epoch: {epoch} train_loss : {loss_b/len(train_loader)}')
  loss_arr.append(loss_b/len(train_loader))
```

```python
model.eval()

test_correct = 0
test_total = 0


with torch.no_grad():
  for x_batch, y_batch in tqdm(test_loader):
    out = model(x_batch)
    _, pred = torch.max(out, axis = 1)

    test_total += y_batch.size(0)
    test_correct += (pred == y_batch).sum().item()

  test_acc = test_correct / test_total
  print(f'Test accuracy is {test_acc *100 :.4f}%\b')
```
